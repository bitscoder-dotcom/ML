```python
#importing the libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder, LabelEncoder
from sklearn.compose import ColumnTransformer

#importing the datasets
dataset = pd.read_csv('E:/user/Documents/ulster/ml/data/Data.csv')
x = dataset.iloc[:, :-1].values
y = dataset.iloc[:, -1].values

# taking care of missing values
imputer = SimpleImputer(missing_values=np.nan, strategy='mean')
imputer.fit(x [:, 1:3])
x[:, 1:3] = imputer.transform(x[:, 1:3])

 # print(x[:, 1:3])

# Encoding categorical data
# Encoding the Independent Variable
ct  =  ColumnTransformer(transformers=[('encoder', OneHotEncoder(), [0])], remainder='passthrough')
x = np.array(ct.fit_transform(x))

print(x)

# Encoding the Dependent Variable
le = LabelEncoder()
y = le.fit_transform(y)

print(y)
```

```python
# Splitting the dataset into the Training set and Test set
from sklearn.model_selection import train_test_split
# x_train ===  the matrix features of the training set
# x_test === the matrix features of the test set
# y_train === the dependent variable of the training set
# y_test === the dependent variable of the test set
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=1 )
```
